```python
# Setup Secrets - Run once and delete after execution
# Fill in the values before runni

# Create Scope
from databricks.sdk import WorkspaceClient

w = WorkspaceClient()
w.secrets.create_scope(scope="etl_sales_scope")
print("Scope created successfully")

# Load secrets
secrets = {
    "STORAGE_ACCOUNT_NAME": "",   # your storage account name
    "STORAGE_ACCOUNT_KEY": "",    # your storage account key
    "AZURE_CONTAINER_NAME": "",         # your container name
    "NEW_DATA_DIRECTORY": "new",
    "PROCESSED_DATA_DESTINATION": "processed",
    "ERROR_DATA_DESTINATION": "error"
}

for key, value in secrets.items():
    w.secrets.put_secret(scope="etl_sales_scope", key=key, string_value=value)
    print(f"Secret {key} saved successfully")
    
print("Delete this notebook after execution")
```
