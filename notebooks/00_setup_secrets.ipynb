```python
# Setup Secrets - Run once and delete after execution
# Fill in the values before running

from databricks.sdk import WorkSpaceClient

# Create Scope
w = WorkSpaceClient()
w.secrets.create_scope(scope="etl-sales-scope")

# Load secrets
secrets = {
    "storage-account-name": "",   # your storage account name
    "storage-account-key": "",    # your storage account key
    "container-name": "",         # your container name
    "new-directory": "new",
    "processed-directory": "processed",
    "error-directory": "error"
}

for key, value in secrets.items():
    w.secrets.put_secret(scope="etl-sales-scope", key=key, string_value=value)
    print(f"Secret {key} saved successfully")
    
print("Delete this notebook after execution")
```
